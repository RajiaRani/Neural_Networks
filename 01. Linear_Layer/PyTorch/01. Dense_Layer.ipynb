# In the code , we will run the Linear layer by using Pytorch

```python
import torch
```

```python
x = torch.rand(100,2, dtype=torch.float32)
print(x[:5])
```

```python
class Linear:
    def __init__(self, n_inputs, n_neurons):
        self.weights = torch.rand(n_inputs, n_neurons, dtype=torch.float32)
        self.bias = torch.zeros((1, n_neurons), dtype=torch.float32)

    def forward(self, inputs):
        self.output = torch.matmul(inputs, self.weights)+self.bias
```

```python
linear1 = Linear(x.shape[1], 6)
linear1.forward(x)
print(linear1.output[:4])

linear2 = Linear(linear1.output.shape[1], 3)
linear2.forward(linear1.output)
print(linear2.output[:4])

linear3 = Linear(linear2.output.shape[1], 1)
linear3.forward(linear2.output)
print(linear3.output[:4])
```
