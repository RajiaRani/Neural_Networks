```python
import numpy as np

# Take 2 inputs and 3 neurons
X = np.array([1, 2])
W = np.array([[0.2, 0.8, -0.5], [0.5, -0.9, 0.26]]) # shape (1,2) 1 sample, 2 features
B = np.array([2, 3, 0.5])

#Forward pss
Z = np.dot(X, W) + B
print("Final result:", Z)
```

ğŸ”¹ Case 2: Weights stored as (neurons Ã— inputs)
2. Neurons Ã— Inputs (this new case)
Shape: (neurons, inputs)

Each row = weights for one neuron

More common in textbooks

Forward pass:

ğ‘ = ğ‘‹ ğ‘Š ğ‘‡ +ğ‘
Z= XWT +b

ğŸ‘‰ Here we must transpose to align shapes.

```python
X = np.array([1,2])
W = np.array([ [0.2, 0.5], [0.8, -0.9], [-0.5, 0.26]])
B = np.array([2, 3, 0.5])
Z = np.dot(X, W.T) + B
print(Z)
```

```python
#Dense layer

class Layer_dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
    X, y = spiral_data(samples=100, classes=3) 

dense1 = Layer_dense(2,3)
dense1.forward(X)
print(dense1.output[:5])

```

# I have 5 inputs - 4 neurons first hidden layer - 3 next - 2 next - 1 neuron final output layer


```python
x = np.random.rand(5, 5)
print(x)
w1 = np.random.rand(4,5)
b1 = np.zeros(4)
print(b1)
print(w1)
```

```python
# 2nd hidden layer having 3 neurons
w2 = np.random.rand(3,4)
b2 = np.zeros(3)
print(w2)
```

```python
# 3rd hidden layer with 2 neurons
w3 = np.random.rand(2, 3)
b3 = np.zeros(2)
print(w3)
```

```python
# final 
w4 = np.random.rand(1, 2)
b4 = np.zeros(1)
print(w4)
```

```python
# Convert lists to numpy array
x_arrays = np.array(x)
w1_array = np.array(w1)
w2_array = np.array(w2)
w3_array = np.array(w3)
w4_array = np.array(w4)
b1_array = np.array(b1)
b2_array = np.array(b2)
b3_array = np.array(b3)
b4_array = np.array(b4)
```

```python
output1 = np.dot(x_arrays, w1_array.T) + b1_array
print(output1)
```

```python
output2 = np.dot(output1, w2_array.T) + b2_array
print(output2)
```

```python
output3 = np.dot(output2, w3_array.T) + b3_array
print(output3)
```

```python
final_output = np.dot(output3, w4_array.T) + b4_array
print(final_output)
```
