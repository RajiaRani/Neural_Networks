```python
import numpy as np
```

```python
# define inputs - 100 rows, 2 columns (100 samples, 2 features)
X = np.random.rand(100, 2)
print(X[:5])
```

```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
     # initialize weights and biases
       self.weights = 0.01 * np.random.rand(n_inputs, n_neurons)
       self.biases = np.zeros((1, n_neurons)) #One bias per neuron

    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
  
        
```

```python
dense1 = Layer_Dense(2, 3)
dense1.forward(X)

dense2 = Layer_Dense(3, 10)
dense2.forward(dense1.output)

dense3 = Layer_Dense(10, 2)
dense3.forward(dense2.output)

dense4 = Layer_Dense(2, 1)
dense4.forward(dense3.output)
```

```python
print(dense4.output[:5])
print(dense4.output.shape)
```

# 10 features with 200 samples in my inputs

```python
X = np.random.rand(200, 10)
print(X[:10])
```

```python
class Dense_Layer:
    def __init__(self, inputs, neurons):
        self.weights = 0.01 * np.random.rand(inputs, neurons)
        self.biases = np.zeros((1, neurons))

    def forward(self, input):
        self.output = np.dot(input, self.weights) + self.biases


dense1 = Dense_Layer(10,20)
dense1.forward(X)

dense2 = Dense_Layer(20, 10)
dense2.forward(dense1.output)

dense3 = Dense_Layer(10,5)
dense3.forward(dense2.output)

dense4 = Dense_Layer(5,3)
dense4.forward(dense3.output)

final_output = Dense_Layer(3,1)
final_output.forward(dense4.output)

# print(dense1.output[:5])
# print(dense1.output.shape)

print(final_output.output[:5])
print(final_output.output.shape)
```
